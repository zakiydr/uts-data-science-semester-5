```python
# Mengubah value dengan LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['jabat'] = le.fit_transform(df['Designation'])
df['tahun'] = le.fit_transform(df['Year'])
```

```python
sns.heatmap(df[['jabat','tahun']].corr(), annot=True)
```

Agar dapat dicari korelasinya, kedua value perlu diubah karena salah satu data merupakan string, dan membuat keduanya memiliki value yang berdekatan
